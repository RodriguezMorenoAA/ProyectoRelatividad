### <span style="color:Blue">Tensor de Faraday para una particula en campo helicoidal</span>

```python
def tensor_de_faraday_campo_helicoidal(B0, B1, x, y, z):
    Ex = Ey = Ez = 0
    Bx = By = 0

    r = np.sqrt(x**2 + y**2 + z**2)
    phi = np.arctan2(y, x)

    Bx = -B1/r * np.sin(phi)
    By = B1/r * np.cos(phi)
    Bz = 0

    F = np.array([[0, -Ex, -Ey, -Ez],
                  [Ex, 0, -Bz, By],
                  [Ey, Bz, 0, -Bx],
                  [Ez, -By, Bx, 0]])
    
    return F
```
