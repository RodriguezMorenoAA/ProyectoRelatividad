### Cuaderno de Clase: Tensores de Riemann, Ricci y Escalar de Curvatura

```python
import einsteinpy.symbolic as es 
import sympy as sp
import numpy as np
from IPython.display import display
```

```python
T, x, y, z = sp.symbols('T x y z')

# Minkowski en cartesianas
gmat = np.diag([1,-1,-1,-1])
g = es.MetricTensor(gmat, [T, x, y, z])

display(g.tensor())
```

```python
Gama = es.ChristoffelSymbols.from_metric(g)
display(Gama.tensor())
```

```python
T, r, teta, z = sp.symbols('T r teta z')

# Minkowski en cilindricas: ds^2 = dT² - dr² - r² dθ² - dz²
gmat = np.diag([1,-1,-r**2,-1])
g = es.MetricTensor(gmat, [T, r, teta, z])
Gama = es.ChristoffelSymbols.from_metric(g)
display(Gama.tensor())
```

```python
Riemann = es.RiemannCurvatureTensor.from_metric(g)
display(Riemann.tensor())
```

Métrica con curvatura:

```python
# Coordenadas esféricas: r, teta, fi
# Elemento de línea espacial: dl^2 = dr^2 + r^2 dθ^2 + r^2 sin^2(θ) dφ^2
# Métrica de Newton: ds^2 = (1 - 2GM/(rc²))dT² - dr^2 - r^2 dθ^2 - r^2 sin^2(θ) dφ^2
T, r, teta, fi, G, M, c = sp.symbols('T r theta phi G M c')

gmat = np.diag([1 - 2*G*M/(r*c**2), -1, -r**2, -r**2*sp.sin(teta)**2])
g = es.MetricTensor(gmat, [T, r, teta, fi])
display(g.tensor())
```

```python
gRiemann_ricci = sp.tensorproduct(ginv.tensor(), 
                                  Riemann_cov.tensor())

Ricci_tensor = sp.simplify(
    sp.tensorcontraction(gRiemann_ricci,
                         (0,2), (1,5)) 
)

display(Ricci_tensor)
```

## Tensor de Ricci de una métrica radialmente simétrica

```python
T, r, teta, fi, G, M, c = sp.symbols('T r theta phi G M c')

A = sp.Function('A')
B = sp.Function('B')

gmat = np.diag([A(r), -B(r), -r**2, -r**2*sp.sin(teta)**2])
g = es.MetricTensor(gmat, [T, r, teta, fi])
display(g.tensor())
```

```python
Ricci = es.RicciTensor.from_metric(g)
display(-Ricci.tensor())
```
